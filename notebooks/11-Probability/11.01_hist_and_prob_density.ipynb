# Histograms and Probability Density

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [1,2,2,2,2,2,3,4,5,6,6,6,6,7,8,9,8,6,3]

plt.hist(data)
plt.xlabel('Value')
plt.ylabel('Number of Apperances (count)')
plt.show()
```

```python
hist_data = np.histogram(data)
print(hist_data)
```

```python
hist_data = np.histogram(data,bins=np.arange(1,11))
print(hist_data)
```

```python
plt.bar(hist_data[1][:-1],hist_data[0])
plt.xlabel('Value')
plt.ylabel('Number of Apperances (count)')
plt.show()
```

```python
# probablitity; sums up to 1
y,x = np.histogram(data,bins=np.arange(1,11))
yP = y/np.sum(y)

plt.bar(x[:-1],yP)
plt.xlabel('Value')
plt.ylabel('Proportion')
plt.xticks(np.arange(1,10))
plt.show()
```

```python
# normalized proportion
y,x = np.histogram(data,bins=np.arange(1,11))
yP = y/np.max(y)

plt.bar(x[:-1],yP)
plt.xlabel('Value')
plt.ylabel('Proportion')
plt.xticks(np.arange(1,10))
plt.show()
```

```python
x = np.linspace(-3,3,1000)

fig,ax = plt.subplots(2,3)
fig.set_size_inches(10,6)

for i in range(3):
    
    y = x**(i+1)
    h = np.histogram(y,50)
    yP = h[0]/np.sum(h[0])
    
    # plot functions
    ax[0,i].plot(x,y)
    ax[0,i].set_title('$y=x^{%g}$' %(i+1))
    
    # plot probability density
    ax[1,i].plot(h[1][:-1],yP)
    
plt.show()
    
```
