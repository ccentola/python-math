# The Central Limit Theorem

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
n = 2000

sig1 = np.sin(np.linspace(0,6*np.pi,n))
sig2 = np.random.rand(n)*2-1

combi = sig1 + sig2

# plot individual signals and their dists
fig,ax = plt.subplots(2,3,figsize=(10,4))

ax[0,0].plot(sig1)
ax[0,0].set_title('Sine Wave')
ax[1,0].hist(sig1,100)

ax[0,1].plot(sig2,'.',alpha=0.3)
ax[0,1].set_title('Uniform Numbers')
ax[1,1].hist(sig2,100)

ax[0,2].plot(combi)
ax[0,2].set_title('Combined Signal')
ax[1,2].hist(combi,100)

for i in np.matrix.flatten(ax):
    i.set_yticks([])
    i.set_xticks([])

plt.show()
```

```python
sns.set()

sns.distplot(combi,100)
plt.show()
```

```python
n = np.arange(50,10000,100)
bins = 50

all_hist = np.zeros((len(n),bins))

for i in range(len(n)):
    sig1 = np.sin(np.linspace(0,6*np.pi,n[i]))
    sig2 = np.random.rand(n[i])*2-1
    
    all_hist[i,:],x = np.histogram(sig1+sig2,bins,density=True)
    

plt.pcolor(x[:-1],n,all_hist,vmax=0.6,cmap='Purples')
plt.xlabel('Data Value')
plt.ylabel('Points in the Sample')
plt.colorbar()

plt.show()
```
