# Joint Probability Distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
n = 1000

x = np.random.randn(n)
y = np.random.randn(n)**2

plt.hist2d(x,y,30,vmax=10)
plt.xlabel('x')
plt.ylabel('$x^2$')

plt.show()
```

```python
sns.jointplot(x,y)
plt.show()
```

```python
sns.jointplot(x,y,kind='kde')
plt.show()
```

```python
sns.jointplot(x,y,kind='hex')
plt.show()
```
