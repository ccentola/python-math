# Building Distrubutions from Random Numbers

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import rankdata
```

```python
shift = 5
stretch = 2
n = 1000

pts = np.random.randn(n)*stretch + shift
print(f'Mean is {np.mean(pts)}, STD is {np.std(pts)}')

# gaussian
fig,ax = plt.subplots(1,2,figsize=(6,3))

ax[0].plot(pts,'s',alpha=0.4)
ax[1].hist(pts,50)

plt.show()
```

```python
shift = 5
stretch = 2
n = 1000

pts = np.random.rand(n)*stretch + shift - 0.5 * stretch
print(f'Mean is {np.mean(pts)}, range is {np.max(pts) - np.min(pts)}')

# uniform
fig,ax = plt.subplots(1,2,figsize=(6,3))

ax[0].plot(pts,'s',alpha=0.4)
ax[1].hist(pts,bins=[4,4.5,5,5.5,6],edgecolor='k')

plt.show()
```

```python
lam = 3.4
n = 1000

# poisson
pts = np.random.poisson(lam,n)

fig,ax = plt.subplots(1,2,figsize=(6,3))
ax[0].plot(pts,'s',alpha=0.4)
ax[1].hist(pts,bins=np.arange(0,np.max(pts)+1),edgecolor='w')

plt.show()
```

## Normalize Distribution to Gaussian

```python
n = 500

orig_data = 2**np.random.randn(n)

# rank
rank_data = rankdata(orig_data)
rank_data = np.append(rank_data,np.max(rank_data)+1)

# normalize
norm_data = 2*(rank_data/np.max(rank_data))-1
norm_data = np.delete(norm_data,n)

# apply inverse hyperbolic function
trans_data = np.arctanh(norm_data)

fig,ax = plt.subplots(1,2)

ax[0].hist(orig_data,30)
ax[0].set_xlabel('value')
ax[0].set_ylabel('count')

ax[1].hist(trans_data,30)
ax[1].set_xlabel('value')
ax[1].set_ylabel('count')

plt.show()
```
