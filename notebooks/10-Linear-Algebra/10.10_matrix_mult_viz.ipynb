# Visualizing Matrix-Vector Multiplication

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = np.array([[0.5,1],[1,0.5]])
v = np.array([1,2])
# v = np.random.randn(2)

Mv = M@v

plt.plot([0,v[0]],[0,v[1]],'b',linewidth=2,label='v')
plt.plot([0,Mv[0]],[0,Mv[1]],'r--',linewidth=2,label='Mv')

plt.axis('square')
plt.axis([-5,5,-5,5])
plt.grid()
plt.legend()
plt.show()
```

```python
vcomp = np.linspace(-2,2,40)

for i in vcomp:
    v = np.array([1,i])
    Mv = M@v
    plt.plot([0,v[0]],[0,v[1]],color=[1-abs(i)/4,0.5,abs(i)/2],alpha=0.8)
    plt.plot([0,Mv[0]],[0,Mv[1]],color=[1-abs(i)/2,abs(i)/4,0.5])
    
plt.axis('square')
plt.axis('off')
plt.show()
```
