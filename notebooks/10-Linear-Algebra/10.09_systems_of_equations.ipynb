# Systems of Equations

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[2,1,-1],[3,4,2],[1,-5,-2]])
y = np.array([1,13,0])

beta1 = np.linalg.inv(X.T@X) @ (X.T@y)
print(beta1)

beta2 = np.linalg.solve(X,y)
print(beta2)
```

```python
X1 = np.array([[3,-1],[-1,1]])
y1 = np.array([6,2])

X2 = np.array([[3,-1],[1.5,-0.5]])
y2 = np.array([6,3])

X3 = np.array([[3,-1],[1.5,-0.5]])
y3 = np.array([6,2])
```

```python
xlim = np.array([0,10])

yy1 = -X1[0,0]/X1[0,1]*xlim + y1[0]/X1[0,1]
yy2 = -X1[1,0]/X1[1,1]*xlim + y1[1]/X1[1,1]


plt.plot(xlim,yy1,'b',label='eq1')
plt.plot(xlim,yy2,'r',label='eq2')

# solve the system
b = np.linalg.solve(X1,y1)
print(b)
plt.plot(b[0],b[1],'ko',markersize=7,label='solution')

plt.legend()
plt.grid()
plt.show()
```

```python
xlim = np.array([0,10])

yy1 = -X2[0,0]/X2[0,1]*xlim + y2[0]/X2[0,1]
yy2 = -X2[1,0]/X2[1,1]*xlim + y2[1]/X2[1,1]


plt.plot(xlim,yy1,'b',label='eq1')
plt.plot(xlim,yy2,'r--',label='eq2')

# solve the system
b = np.linalg.solve(X2,y2)
print(b)
plt.plot(b[0],b[1],'ko',markersize=7,label='solution')

plt.legend()
plt.grid()
plt.show()
```
