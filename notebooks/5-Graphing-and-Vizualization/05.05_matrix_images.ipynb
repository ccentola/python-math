# Creating Images from Matrices

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = [[1,2],[1,4]]

plt.imshow(a)

plt.xticks([0,1])
plt.yticks([0,1])

plt.show()
```

```python
a = np.zeros((10,14))

# for i in range(0,np.shape(a)[0]):
#     for j in range(0,np.shape(a)[1]):
#         a[i,j] = 3*i - 4*j
        
# for i in range(0,np.shape(a)[0]):
#     for j in range(0,np.shape(a)[1]):
#         plt.text(
#             j,i,int(a[i,j]),
#             horizontalalignment='center',
#             verticalalignment='center'
#         )

for i in range(0,np.shape(a)[0]):
    for j in range(0,np.shape(a)[1]):
        
        a[i,j] = 3*i - 4*j
        
        plt.text(
            j,i,int(a[i,j]),
            horizontalalignment='center',
            verticalalignment='center'
        )

plt.set_cmap('twilight_shifted')
plt.imshow(a)
```

```python
n = 10

c = np.zeros((n,n))

for i in range(0,n):
    for j in range(0,n):
        
        c[i,j] = (-1)**(i+j)
        

plt.set_cmap('Greys')
plt.tick_params(labelleft=False,labelbottom=False)
plt.imshow(c)
```
