# Graphing Lines

```python
import matplotlib.pyplot as plt
import sympy as sym
from math import sqrt
import numpy as np
```

```python
p1 = [-3,-1]
p2 = [4,4]

plt.plot([p1[0],p2[0]],[p1[1],p2[1]])

plt.show()
```

```python
x = 3
y = 5

plt.plot(x,y,'ro')
plt.plot([0,x],[0,y],'r')

plt.axis('square')
plt.axis([-6,6,-6,6])
plt.grid()

plt.plot([-6,6],[0,0],'k')
plt.plot([0,0],[-6,6],'k')

plt.show()
```

```python
x = sym.symbols('x')

y = abs(x)**(1/2)

for i in range(-20,21):
    plt.plot([0,i],[0,y.subs(x,i)])
    
plt.show()
```

## Slope-Intercept Form

```python
# y = mx + b

x = [-5,5]
m = 2
b = 1

y = [340,2345]

for i in range(0,len(x)):
    y[i] = m*x[i] + b
    
plt.plot(x,y,label=f'y={m}x+{b}')

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca() # get current axis
plt.plot(axis.get_xlim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--')
plt.legend()
    
plt.show()
```

```python
# y = mx + b

x = [-5,5]
m = 2
b = 1

# y = [340,2345]

# for i in range(0,len(x)):
#     y[i] = m*x[i] + b

y = m*np.array(x) + b
    
plt.plot(x,y,label=f'y={m}x+{b}')

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca() # get current axis
plt.plot(axis.get_xlim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--')
plt.legend()
    
plt.show()
```

```python
# y = mx + b

x = [-5,5]
m = [.7,-1.25]
b = [-2,.75]

# y = [340,2345]

# for i in range(0,len(x)):
#     y[i] = m*x[i] + b

for i in range(0,len(m)):
    y = m[i]*np.array(x) + b[i]
    plt.plot(x,y,label=f'y={m[i]}x+{b[i]}')

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca() # get current axis
plt.plot(axis.get_xlim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--')
plt.legend()
    
plt.show()
```
