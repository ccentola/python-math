# Find a Point on a Gaussian

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = .999
t = np.linspace(-3,3,1001)

#
g = np.exp(-4*np.log(2)*t**2 / h**2)

plt.plot(t,g)
plt.xlabel('t')
plt.ylabel('g')
plt.title(f'Gaussian with FWHM={h}')

plt.show()
```

```python
N = 100
t = np.linspace(-3,3,1001)
h = np.zeros(N)

G = np.zeros((N,len(t)))

for i in range(N):
    h[i] = (i+1)/N
    G[i,:] = np.exp(-4*np.log(2)*t**2 / h[i]**2)
    
plt.pcolormesh(t,h,G)
plt.xlabel('t')
plt.ylabel('h')
plt.show()
```
