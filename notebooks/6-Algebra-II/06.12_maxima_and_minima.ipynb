# Maxima and Minima

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.signal import find_peaks
```

```python
l = [-4,5.5,1,2]

max_val = np.max(l)
max_loc = np.argmax(l)

print(max_val)
print(max_loc)

l[max_loc]
```

```python
x = np.linspace(0,2**np.pi,20)
f = -(np.cos(x) + np.sqrt(x))

max_val = np.max(f)
max_loc = np.argmax(f)

min_val = np.min(f)
min_loc = np.argmin(f)

plt.plot(x,f,'o-', color='gray')
plt.plot(x[max_loc],max_val,'rs')
plt.plot(x[min_loc],min_val,'bs')

plt.show()
```

```python
x = np.linspace(0,12*np.pi,213)
f = -np.cos(x) - np.sqrt(x)

peaks = find_peaks(f)

plt.plot(x,f, color='gray')
plt.plot(x[peaks[0]],f[peaks[0]], 'o')

plt.show()
```
