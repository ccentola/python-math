# Even and Odd Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,20)

f_even = x**2
f_even_neg = (-x)**2

plt.plot(x,f_even)
plt.plot(x,f_even_neg,'ro')

plt.show()
```

```python
x = np.linspace(-5,5,20)

f_odd = x**3
f_odd_neg = (-x)**3
f_neg_odd = -f_odd

plt.plot(x,f_odd)
plt.plot(x,f_odd_neg,'ro')
plt.plot(x,f_neg_odd,'g')

plt.show()
```
