# Graphing Function Tangent Lines

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
x = sym.symbols('x')

f = x**2
d = sym.diff(f)

xa = 1

fa = f.subs(x,xa)
da = d.subs(x,xa)

xx = np.linspace(-2,2,200)
f_fun = sym.lambdify(x,f)(xx)
d_fun = sym.lambdify(x,d)(xx)

tanline = da * (xx-xa) + fa

plt.plot(xx,f_fun,label='f(x)')
plt.plot(xx,tanline,label='tangent')
plt.plot(xa,fa,'ro')

plt.axis('square')
plt.axis([-2,2,-2,2])

ax = plt.gca()
plt.plot(ax.get_xlim(),[0,0],'k--')
plt.plot([0,0],ax.get_ylim(),'k--')
plt.legend()

plt.show()
```

```python
def compute_tangent(f,xa,bnds):
    
    df = sym.diff(f)
    fa = f.subs(x,xa)
    dfa = df.subs(x,xa)
    
    xx = np.linspace(bnds[0],bnds[1],200)
    return dfa * (xx-xa) + fa
```

```python
x = sym.symbols('x')
f = x**2
xx = np.linspace(-2,2,200)
ffun = sym.lambdify(x,f)(xx)

for a in xx:
    y = compute_tangent(f,a,xx[[0,-1]])
    plt.plot(xx,y,color=[abs(a)/3,abs(a)/4,abs(a)/2])

plt.plot(xx,ffun)
plt.axis('square')
plt.axis([-2,2,-1,3])
plt.axis('off')
plt.show()

```
