# Piecewise Functions

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
from sympy.abc import x

piece1 = 0
piece2 = -2*x
piece3 = x**3/10

fx = sym.Piecewise( 
    (piece1,x<0),
    (piece2,(x>=0) & (x<3)),
    (piece3,x>=3)
)

fxx = sym.lambdify(x,fx)
xx = np.linspace(-3,5,1234)

plt.plot(xx,fxx(xx))
plt.show()
```

```python
from sympy.abc import x

piece1 = x**3
piece2 = sym.log(x,2)

fx = sym.Piecewise( 
    (piece1,x<=0),
    (piece2,x>0)
)

fxx = sym.lambdify(x,fx)
xx = np.linspace(-3,5,1234)

display(Math('f(x) = ' + sym.latex(fx)))

plt.plot(xx,fxx(xx),'k')
plt.xlim([-2,2])
plt.ylim([-10,3])

plt.show()
```
