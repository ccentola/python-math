# Random Numbers

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nums = np.random.rand()
nums
```

```python
nums = np.random.rand(1000)

plt.plot(nums,'s')
plt.show()

plt.hist(nums)
plt.show()
```

```python
nums = np.random.rand(1000)

nums = nums + 3

plt.plot(nums,'s')
plt.show()

plt.hist(nums)
plt.show()
```

```python
nums = np.random.rand(1000)

min_val = 2
max_val = 17

nums = nums*(max_val-min_val) + min_val

plt.plot(nums,'s')
plt.show()

plt.hist(nums)
plt.show()
```

```python
# normal dist
nums = np.random.randn(1000)

min_val = 2
max_val = 17

nums = nums*(max_val-min_val) + min_val

plt.plot(nums,'s')
plt.show()

plt.hist(nums)
plt.show()
```

```python

m = 15
sd = 4.3

# normalize to get exact values
nums = np.random.randn(1000) * sd + m
nums = nums - np.mean(nums)
nums = nums/np.std(nums)
nums = nums * sd + m

plt.plot(nums,'s')
plt.show()

plt.hist(nums)
plt.show()

print(f'The distribution has a mean of {np.mean(nums)} and an sd of {np.std(nums)}')
```

## Plotting Random Phase Angles

```python
angs = np.random.randn(100)

for i in angs:
    plt.plot([0,np.cos(i)],[0,np.sin(i)])
    
plt.axis('square')
plt.axis([-1,1,-1,1])
plt.axis('off')
plt.show()
```

```python
angs = np.random.randn(100) * 2 * np.pi

for i in angs:
    plt.plot([0,np.cos(i)],[0,np.sin(i)])
    
plt.axis('square')
plt.axis([-1,1,-1,1])
plt.axis('off')
plt.show()
```
