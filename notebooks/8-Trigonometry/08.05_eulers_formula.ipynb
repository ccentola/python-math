# Euler's Formula

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = np.pi / 4
m = 2.3

eul = m * np.exp(1j*k)
cis = m * (np.exp(k) + 1j * np.sin(k))

print(cis)
print(eul)
```

```python
mag = np.abs(eul)
ang = np.angle(eul)

plt.polar([0,ang],[0,mag],'r')
plt.polar(k,m,'bo')

plt.show()
```

```python
def euler_from_cos_sin():
    
    re = eval(input('cosine part: '))
    im = eval(input('sine part: '))
    
    m = np.sqrt(re**2 + im**2)
    k = np.arctan2(im,re)
    
    plt.polar([0,k],[0,m])
    plt.title('me$^{i\\phi}$, m=%g, $\\phi$=%g' %(m,k))
    plt.thetagrids([0,45,130,200,222.2])
    
    plt.show()
    
    
```

```python
euler_from_cos_sin()
```

```python
# random exploding euler
n_vects = 200
ms = np.random.rand(n_vects)
ks = np.random.rand(n_vects)*2*np.pi

for i in range(0,n_vects):
    r = np.random.rand()
    
    if r < 0.4:
        clr = [1, 0.2, 0.7]
    elif r > 0.4 and r < 0.8:
        clr = [0.7, 0.2, 1]
    else:
        clr = [0,1,0]
        
    
    plt.polar([0,ks[i]],[0,ms[i]],color=clr)
    
plt.axis('off')
plt.show()
```

```python
t = np.linspace(0,8*np.pi,1000)

r1 = np.random.rand()
r2 = np.random.rand()

x = np.cos(r1*t)
y = np.sin(r2*t)

# x = np.append(x,x[0])
# y = np.append(y,y[0])

plt.plot(x,y,'k')
plt.axis('square')
plt.axis([-1.1,1.1,-1.1,1.1])
plt.axis('off')
plt.title(f'$r_1$={np.round(r1,2)}, $r_2$={np.round(r2,2)}')

plt.show()
```
