# Graphing Resolution for Sine, Cosine, and Tangent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,6*np.pi,400)

plt.plot(x,np.sin(x),'r',label='$\\sin(\\theta)$')
plt.plot(x,np.cos(x),'b',label='$\\cos(\\theta)$')

plt.legend()
plt.xlabel('radians')
plt.ylabel('function value')
plt.show()
```

```python
th = np.linspace(0,4*np.pi,300)

plt.plot(th, np.tan(th),'k')
plt.ylim([-5,5])
plt.show()
```

```python
x = np.linspace(-2*np.pi,2*np.pi,200)

y1 = np.sin(x + np.cos(x))
y2 = np.cos(x + np.sin(x))

plt.plot(x,y1,x,y2)
plt.show()
```

```python
theta = np.pi*7/6

# plot the unit circle
x = np.linspace(0,2*np.pi,200)
plt.plot(np.cos(x),np.sin(x),color='gray')
plt.plot([-1.1,1.1],[0,0],'--',color='gray')
plt.plot([0,0],[-1.1,1.1],'--',color='gray')

# plot the arc
plt.plot(
    np.cos(np.linspace(0,theta)),
    np.sin(np.linspace(0,theta)),
    'k',
    linewidth=3
)

plt.plot([0,np.cos(theta)],[0,np.sin(theta)],'k:')
plt.plot(np.cos(theta),np.sin(theta),'ko')

# plot sine/cosine components
plt.plot([0,0],[0,np.sin(theta)],'r')
plt.plot([0,np.cos(theta)],[np.sin(theta),np.sin(theta)],'r:')

plt.plot([0,np.cos(theta)],[0,0],'b')
plt.plot([np.cos(theta),np.cos(theta)],[0,np.sin(theta)],'b:')

plt.gca().set_aspect('equal')
plt.title(f'$\\theta = {theta}$')
plt.show()
```
