# Trigonometric Art

## Astroid Radial Curve

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-6*np.pi,6*np.pi,1000)
a = t

x = a * np.cos(t)**3
y = a * np.sin(t)**3

plt.plot(x,y,'m',linewidth=3)
plt.axis('off')
plt.axis('square')
plt.show()
```

```python
n = list(range(0,9))
t = np.linspace(-6*np.pi,6*np.pi,1000)

fig,ax = plt.subplots(3,3,figsize=(10,6))
row,col = np.indices((3,3))

for i in n:
    
    a = t**i

    x = a * np.cos(t)**3
    y = a * np.sin(t)**3
    
    r = row.ravel()[i]
    c = col.ravel()[i]
    
    ax[r,c].plot(x,y,'k')
    ax[r,c].axis('off')
    ax[r,c].axis('square')
    
plt.show()
```

## Rose Curves

```python
k = 2
t = np.linspace(0,8*np.pi,1000)

x = np.cos(k*t)*np.cos(t)
y = np.cos(k*t)*np.sin(t)

plt.plot(x,y,'m',linewidth=3)
plt.axis('off')
plt.axis('square')
plt.show()
```

```python

t = np.linspace(0,8*np.pi,1000)

n = list(range(0,9))

fig,ax = plt.subplots(3,3,figsize=(10,6))
row,col = np.indices((3,3))

for i in n:
    
    k = i/2

    x = np.cos(k*t)*np.cos(t)
    y = np.cos(k*t)*np.sin(t)
    
    r = row.ravel()[i]
    c = col.ravel()[i]
    
    ax[r,c].plot(x,y,'k',linewidth=0.5)
    ax[r,c].set_title(f'k={k}')
    ax[r,c].axis('off')
    ax[r,c].axis('square')
    
plt.show()
```

## Squircle

```python
x = np.linspace(-1,1,2001)
y = (1 - x** 4)**(1/4)

plt.plot(x,y,'m',linewidth=3)
plt.plot(x,-y,'m',linewidth=3)

plt.axis('off')
plt.axis('square')
plt.show()
```

## Logartithmic Sprial

```python
t = np.linspace(0,8*np.pi,1000)
k = -3

x = np.cos(t) * np.exp(t/k)
y = np.sin(t) * np.exp(t/k)

plt.plot(x,y,'m',linewidth=3)

plt.axis('off')
plt.axis('square')
plt.show()
```

```python
t = np.linspace(0,8*np.pi,1000)

n = 150
ks = np.linspace(-10,-2,n)


for i in range(n):
    
    x = np.cos(t) * np.exp(t/ks[i])
    y = np.sin(t) * np.exp(t/ks[i])
    
    plt.plot(x,y,color=[0,i/n,i/n])
    
    

plt.axis('off')
plt.axis('square')    
plt.show()
```

## Logistic Map

```python
rs = np.linspace(1,4,1000)

N = 500
x = 0.5 + np.zeros(N)
endcap = np.arange(round(N*0.9),N)

for ri in range(len(rs)):
    
    for n in range(N-1):
        
        x[n+1] = rs[ri] * x[n] * (1 - x[n])
        
    u = np.unique(x[endcap])
    r = rs[ri]*np.ones(len(u))
    plt.plot(r,u,'.',markersize=1)
    

plt.show()
    
```
